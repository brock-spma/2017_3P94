```python
# load csv to dataframe
# set index
dm = pandas.read_csv('shots.csv')
dm = dm.set_index(['season', 'gamenumber', 'eventnumber']) 
type(dm)
```

### Columns

```python
dt = dm.copy()
dt = dm[['gamedate', 'vteamcode', 'hteamcode']]
dt.head()
```

```python
# drop column
dt = dm.copy()
dt = dm.drop(['playernumber', 'namegoalie'], axis=1)
dt.head()
```

```python
# create columns (difference in goals scored and goals allowed)
dt = dm.copy()
dt['dgoals'] = dt.tgoals - dt.ogoals
dt.head()
```

### Index (rows)

```python
# cross section
dt = dm.copy()
dt = dt.xs(key=20002, level='gamenumber', drop_level=False, axis=0)
dt.head()
```

```python
dt = dm.copy()
dt = dt.xs(key=[20002, 10], level=['gamenumber', 'eventnumber'], drop_level=False, axis=0)
dt.head()
```

```python
# subset data based on index
dt = dm.copy()
dt = dt.query('gamenumber>20002')
dt.head()
```

```python
# subset data based on columns
dt = dm.copy()
dt = dt[dt.period>2]
dt.head()
```

```python
# append column to index
dt = dm.copy()
dt = dt.set_index(['vteamcode', 'hteamcode', 'isTHome', 'period'], append=True)
dt.head()
```
